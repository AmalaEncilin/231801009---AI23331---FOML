```python
import numpy as np
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])
W = np.random.randn(2)
b = np.random.randn()
alpha = 0.1
epochs = 10000
for epoch in range(epochs):
    for i in range(X.shape[0]):
        z = np.dot(X[i], W) + b
        y_pred = 1 / (1 + np.exp(-z))
        error = y[i] - y_pred
        W += alpha * error * X[i]
        b += alpha * error
for i in range(X.shape[0]):
    z = np.dot(X[i], W) + b
    y_pred = 1 / (1 + np.exp(-z))
    print(f"Input: {X[i]}, Predicted Output: {y_pred}")
```
